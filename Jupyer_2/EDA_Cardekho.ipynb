```python
fig = px.box(df, x = 'oem', y = 'price', color = 'oem')

fig.update_layout(
    height = 500, width = 1200,
    title_text = "Brand by Price",  # title of plot
    xaxis_title_text = "oem",  # x-axis label
    yaxis_title_text = "price",  # y-axis label
)

fig.show()
```
