```python
def specification(data):
    extract = {}
    for value in data['top']:
        extract[value['key']] = value['value']

    try:
        for i in range(3):
            for value in data['data'][i]['list']:
                extract[value['key']] = value['value']
    except:
        pass

    return extract

result_df.new_car_specs = result_df.new_car_specs.apply(lambda x: specification(eval(x)))
speci_df=pd.DataFrame(result_df.new_car_specs.tolist())
speci_df
```
